# TechGadgets Customer Support Chatbot

## Overview
This chatbot assists customers by answering questions about TechGadgets Inc.’s products, shipping, return policy, and customer support. The chatbot integrates sentiment analysis to adjust its tone based on the user’s mood.

## Features
- **Sentiment Analysis**: Determines whether the user’s question is positive, neutral, or negative.
- **Context-Aware Responses**: Answers questions based on the business context.
- **Interactive Interface**: Gradio-powered interface that allows users to type questions and receive answers in real-time.

## Sentiment Analysis
The sentiment analysis is powered by Hugging Face’s pipeline

## How It Works
1. The user asks a question.
2. Sentiment analysis is performed to detect whether the question is positive, negative, or neutral.
3. The chatbot generates a response based on the sentiment and business context.

```python
# Load sentiment analysis and question answering models
sentiment_analyzer = pipeline("sentiment-analysis")
qa_pipeline = pipeline("question-answering")
```

```python
# Business context
business_context = """
TechGadgets Inc. is a leading electronics retailer specializing in smartphones, laptops, and accessories.

Products:
- Smartphones: We offer the latest models from Apple, Samsung, and Google. Prices range from $300 to $1500.
- Laptops: We carry brands like Dell, HP, and Lenovo. Prices range from $500 to $2500.
- Accessories: We stock chargers, cases, and screen protectors for all our devices.

Shipping Policy:
- Free standard shipping on orders over $50.
- Standard shipping (5-7 business days) costs $5.99 for orders under $50.
- Express shipping (2-3 business days) is available for an additional $15.

Return Policy:
- 30-day return policy for unopened items.
- 14-day return policy for opened items in original condition.
- Refunds are processed within 5-7 business days after receiving the returned item.

Customer Support:
- Phone support available Monday to Friday, 9 AM to 6 PM EST.
- Email support available 24/7 with responses typically within 24 hours.

Warranty:
- All products come with a standard 1-year manufacturer warranty.
- Extended warranties are available for purchase, covering up to 3 years.
"""
```

```python
# function to generate sentiments and reponse based on sentiment
def get_sentiment(text):
    result = sentiment_analyzer(text)[0]

    return result['label'], result['score']

def generate_response(question):
    # Use the QA pipeline to get an answer based on the business context
    answer = qa_pipeline(question=question, context=business_context)

    # get the sentiment. whether posive or nagative
    sentiment_label, _ = get_sentiment(question)

     # Adjust the response based on sentiment
    if sentiment_label == "POSITIVE":
        response = f"I'm glad I can help! 😊 Here's the information: {answer['answer']}"
    elif sentiment_label == "NEGATIVE":
        response = f"I'm sorry to hear you're having trouble. 😔 Let me help clarify: {answer['answer']}"
    else:  # Neutral sentiment
        response = f"Here's the information you requested: {answer['answer']}"

    return response
```

```python
# Create the Gradio interface
interface = gr.Interface(
    fn=generate_response,
    inputs=gr.Textbox(lines=2, label="Your Question"),
    outputs=gr.Textbox(label="Chatbot Response"),
    title="TechGadgets Customer Support Chatbot",
    description="Ask a question about TechGadgets products, shipping, returns, or support, and our AI assistant will help you.",
    examples=[
        ["What smartphones do you offer?"],
        ["How long does shipping take?"],
        ["What's your return policy?"],
        ["Do you offer warranties on your products?"]
    ]
)

interface.launch()
```
